### Recommendation Engine :

The Recommendation system/engine is used for Recommending the items,movies,songs,food,books and many more this all are recommended based on the past activities or the interests of the human.The recommendation system will understand the human nature and likelihood of human based on that it will recommend the products, food,movies,music and many more.

###### The Recommendation System is of 2 types :

###### 1 . Content-Based Filtering : 

Content based recommendation will analyse the item similarity based on that it will going to recommend which of the items are having close matchs that will going recommend to the persons who purchased/watched/listened similar feature item/movie/song.....

###### 2.Collaborative-Based Filtering

Collaborative  filtering will going to recommend based on the similar interests.It will going to gather all the persons with similar interests and which ever one person buy/watch it will going to recommended to all with the similar interests.

##### Movie Recommendation 

Movies data vectorization:

```python

from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
features=tf.fit_transform(movies['description'])
print(features)
```

Movies cosine similarities :

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(features)
print(similarity)
```

shows cosine similarities:

List of all the movies in the dataset :

```python
list_of_movies=movies['title'].tolist()
list_of_movies
```

```python
movie1=input('Enter movie name :')
```

```python
import difflib
```

```python
close_match=difflib.get_close_matches(movie1,list_of_movies)
print(close_match)
```

```python
close_match=close_match[0]
close_match
```

```python
index_of_the_movie=movies[movies.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similarity_score=sorted(similarity_score,key=lambda x:x[1],reverse=True)
sorted_similarity_score
```

Print Name of the similar movies based on the index :

```python
print("Movie suggested for you:/n")
i=1
for movie in sorted_similarity_score:
    index=movie[0]
    title_from_index=movies[movies.index==index]['title'].values[0]
    if (i<=10):
        print(i,'.',title_from_index)
        i+=1
```

##### Shows Recommendation :

```python
from sklearn.feature_extraction.text import TfidfVectorizer
t2=TfidfVectorizer()
values_of_shows=t2.fit_transform(shows['description'])
print(values_of_shows)
```

cosine_similaritys of shows

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity_shows=cosine_similarity(values_of_shows)
cosine_similarity_shows
```

All the shows in a list

```python
list_of_shows=shows['title'].tolist()
```

```python
show1=input("Enter a show name:")
```

```python
import difflib
```

```python
close_shows=difflib.get_close_matches(show1,list_of_shows)
close_shows
```

```python
close_shows=close_shows[0]
close_shows
```

```python
index_of_the_shows=shows[shows.title==close_shows]['index'].values[0]
index_of_the_shows
```

```python
similarity_score_of_shows=list(enumerate(cosine_similarity_shows[index_of_the_shows]))
similarity_score_of_shows
```

Print name of the similar shows based on the index numbers

```python
print("shows recommeded for you are:/n")
j=1
for show in sorted_similarity_score_shows:
    index=show[0]
    title_from_shows=shows[shows.index==index]['title'].values[0]
    if j<=10:
        print(j,".",title_from_shows)
        j+=1
```
